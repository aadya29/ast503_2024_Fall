# Q1. The $k$-correction in the sub-mm and the optical (60 pts)

You know how to calculate the flux synthetically in some band, but what about comparing the flux of the same class of objects at different redshifts. Necessarily, you'll have to compare the rest-frame flux with the observer frame flux further in the red at some redshift $z$ because the Universe is expanding and frequency of photons is shifting. 

To do this, you need the "$k$-correction" which is nicely summarized in Hogg et al. 2002: https://arxiv.org/abs/astro-ph/0210394
(You might start by looking at Eqn. 12, and comparing it to the synthetic integrals for the flux you implemented in Homework 1)

As we discussed in class (or as you might know if you've listened to a talk from Prof. Joaquin Vieira) galaxies have an interesting effect where the flux density $F_\nu$ from about 250$\mu m$ becomes roughly constant over a large range in redshift because you climb the Rayleigh-Jeans tail as you increase redshift, which compensates for the decrease in the flux density because of increasing distance. In the optical, you have no such luck, and indeed, depending on the class of galaxy, the $k$-correction means that galaxies in a fixed observer band might seem even fainter than you might expect from the inverse square law. 

In particular, the $k$-correction goes negative.

````python
# function to get redshifted flux density in mJy from the SED
def get_fnu_z(z, wave, fnu):
    wave_obs = wave * (1+z)
    fnu_obs = fnu * (1+z)
    return wave_obs, fnu_obs

#plot the SED at different redshifts
plt.figure(figsize=(10,6))
for i in z:
    plt.plot(get_fnu_z(i, data['wave'], data['fnu']), label='z='+str(i))
plt.xlabel('Wavelength (microns)')
plt.ylabel('Fnu (mJy)')
plt.title('SED at different redshifts')
plt.legend()
plt.xscale('log')
# plt.yscale('log')
plt.show()
````

We want to understand *how* brightness limits affect the use of type Ia supernovae observations in the measurement of the Hubble constant, $H_0$. To do so, we will use Monte Carlo computations to generate a set of simulated data. 

A subtle effect occurs in this measurement, called Malmquist bias, that can affect the result. The effect is caused by the range of apparent brightness for supernova. Supernova in our simulation have an absolute magnitude of M=-19.

````python
# make the above code into functions

def generate_distances(N, radius):
    #Find z for 2500 Mpc
    z_max = z_at_value(cosmo.luminosity_distance, radius*u.Mpc) #max redshift for 2500 Mpc
    #generate random redshifts
    z = np.random.uniform(0,z_max, N)
    #calculate distances
    distances = cosmo.luminosity_distance(z).to(u.Mpc)
    return distances

def generate_luminosities(N, M):
    #sample N points from a normal distribution
    lum = np.random.normal(0, 1, N) + M
    return lum

def generate_apparent_magnitudes(lum, distances):
    #convert to apparent magnitude
    m = lum + 5*np.log10(distances.to(u.pc).value) - 5
    return m

def generate_recessional_velocities(distances):
    #get recessional velocities for the SNe
    v = cosmo.H0 * distances
    #add random peculiar velocities
    v_p = np.random.normal(0, 600, N)*u.km/u.s
    #total velocity
    v_total = v + v_p
    return v_total

def generate_observed_distances(m, M):
    #find the observed distance by assuming M = -19 and H0 = 72 km/s/Mpc
    distances_observed = 10**((m - M)/5 + 1)*u.pc
    return distances_observed

def plot_apparent_magnitude_vs_distance(distances, m, m_21):
    #plot the m vs distance for all SNe
    plt.figure(figsize=(10,6))
    plt.scatter(distances, m, color='blue', label='All SNe')
    plt.scatter(distances[m<21], m_21, color='red', label='Detected SNe')
    plt.axhline(y=21, color='black', linestyle='--', label='Detection limit')
    plt.xlabel('Distance (Mpc)')
    plt.ylabel('Apparent magnitude')
    plt.title('Apparent magnitude vs distance')
    plt.legend()
    plt.show()

def plot_total_velocity_vs_distance(distances, v_total, distances_observed, m_21):
    #plot the total velocity vs distance
    plt.figure(figsize=(10,6))
    plt.scatter(distances, v_total, color='blue', label='True distances')
    plt.scatter(distances_observed.to(u.Mpc), v_total[m<21], color='red', label='Observed distances')
    plt.axvline(x=1000, color='black', linestyle='--', label='Detection Limited distance')
    plt.xlabel('Distance (Mpc)')
    plt.ylabel('Total velocity (km/s)')
    plt.title('Total velocity vs distance')
    plt.legend()
    plt.show()

def calculate_H0(distances_observed, v_total):
    # calculate the Hubble constant using linear regression
    slope, intercept, r_value, p_value, std_err = linregress(distances_observed.to(u.Mpc).value, v_total[m<21].value)
    H0 = slope
    return H0
````

````python
#redo the above analysi using the functions for N=1000 and N =10000
N = 1000
radius = 2500 #Mpc

distances = generate_distances(N, radius)
mean_distance = np.mean(distances)
std_distance = np.std(distances)

M = -19
lum = generate_luminosities(N, M)
m = generate_apparent_magnitudes(lum, distances)
m_21 = m[m<21]

v_total = generate_recessional_velocities(distances)
distances_observed = generate_observed_distances(m_21, M)

plot_apparent_magnitude_vs_distance(distances, m, m_21)
plot_total_velocity_vs_distance(distances, v_total, distances_observed, m_21)

H0 = calculate_H0(distances_observed, v_total)
print('H0:', H0)

````
